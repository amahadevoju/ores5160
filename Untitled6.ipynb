```python
data.drop(columns=['Party', 'SDOHi 2019', 'Percentage of population infected', 'COVID-19 Cases per capita (As of 09/17/2021)', 'Population Density (2020)', 'Total cases rank',
       'Per Capita Rank', 'Population density rank', 'Case vs density difference', 'Per capita vs density difference', 'Well-Being Index 2019'], axis=1, inplace=True)
```

```python
data.columns
```

```python
data = data.rename(columns ={'State': 'State',  'Total State Population (2021)': 'Population_2021', 'Total COVID-19 Cases (As of 09/17/2021)': 'Total_Cases_09/17/2021', 'COVID-19 Deaths': 'Total_COVID19_Deaths', 'Death Rate': 'Death_Rate', 'Mask Mandate Start': 'Mask_Mandate_Start_Date', 'Mask Mandate End': 'Mask_Mandate_End_Date', 'Total Mask Days': 'Total_Mask_Days', 'Vaccination Rate (1 dose)': 'Vaccination Rate_1_dose', 'Vaccination Rate (2 dose)': 'Vaccination Rate_2_dose', 'Vaccine Rank': 'Vaccine Rank'})
data
```

```python
data.dropna()
```

```python
data.drop_duplicates()
```

```python
sns.boxplot(data['Total_COVID19_Deaths'])
```

```python
upper_limit = data['Total_COVID19_Deaths'].mean() + 3 * data['Total_COVID19_Deaths'].std()
lower_limit = data['Total_COVID19_Deaths'].mean() - 3 * data['Total_COVID19_Deaths'].std()
print('upper limit:', upper_limit)
print('lower limit:', lower_limit)
```

```python
data.loc[(data['Total_COVID19_Deaths'] > upper_limit) | (data['Total_COVID19_Deaths'] < lower_limit)]
```

```python
new_data = data.loc[(data['Total_COVID19_Deaths'] < upper_limit) & (data['Total_COVID19_Deaths'] > lower_limit)]
print('before removing outliers:', len(data))
print('after removing outliers:', len(new_data))
print('outliers:', len(data)-len(new_data))
```

```python
sns.boxplot(data['Population_2021'])
```

```python
upper_limit = data['Population_2021'].mean() + 3 * data['Population_2021'].std()
lower_limit = data['Population_2021'].mean() - 3 * data['Population_2021'].std()
print('upper limit:', upper_limit)
print('lower limit:', lower_limit)
```

```python
data.loc[(data['Population_2021'] > upper_limit) | (data['Population_2021'] < lower_limit)]
```

```python
new_data1 = data.loc[(data['Population_2021'] < upper_limit) & (data['Population_2021'] > lower_limit)]
print('before removing outliers:', len(data))
print('after removing outliers:', len(new_data1))
print('outliers:', len(data)-len(new_data1))
```

```python
def to_int64(col):
    return pd.to_numeric(col, errors='coerce').astype('Int64')
data['Total_Mask_Days'] = to_int64(data['Total_Mask_Days'])
```
